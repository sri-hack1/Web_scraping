```python
# Modules used
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Creating headers
url = "https://www.cnbc.com/world/?region=world"
headers = {
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36"
           }

# Getting response
response = requests.get(url, headers = headers)
```

```python
# Parsing HTMl content
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Scrap heading
headings = soup.find_all("a", class_ = "LatestNews-headline")
heading = [heading.text.strip() for heading in headings]

# Date
dates = soup.find_all("time", class_ = "LatestNews-timestamp")
date = [date.text.strip() for date in dates]

# News link
links = soup.find_all("a", class_= "LatestNews-headline")
link = [link["href"] for link in links]
```

```python
# Creating dataframe
new_df = pd.DataFrame({
    "Heading": heading,
    "Dates":date,
    "Links":link
})
new_df
```
