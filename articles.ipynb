```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/"
headers = {
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36"
           }
response = requests.get(url, headers = headers)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
#  Scraping titles
titles_raw = soup.find_all("h2", class_= "h5 article-title")
titles = [title.text.strip() for title in titles_raw]

# Scraping dates
dates_raw = soup.find_all("p", class_= "article-date")
dates =[date.text.strip() for date in dates_raw]

# Scraping authors name
authors_raw = soup.find_all("p", class_="article-authors")
authors = [author.text.strip() for author in authors_raw]
```

```python
# Creating DataFrame
df = pd.DataFrame({
    "Titles":titles,
    "Dates":dates,
    "Authors":authors
})
df
```
